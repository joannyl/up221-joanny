### Maps :)

Using data from LA County's GIS portal in geojson format to map

```python
import geopandas as gpd
```

```python
tracts=gpd.read_file('Data/Census_Tracts_2020.geojson')
tracts.head()
```

Testing out the map 

```python
tracts.plot(figsize=(12,10))
```

```python
tracts.info(verbose=True, show_counts=True)
```

```python
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
# create a FIPS column
tracts['FIPS'] ='06' + '037' + tracts['CT20']
```

Merging data to map census data from social explorer 

+ Creating a new dataframe 

```python
tracts_age=tracts.merge(df2,on="FIPS")
```

```python
tracts_age.head()
```

Now to make some maps :)

```python
tracts_age.plot(figsize=(12,10),
                 column='TotalPop <5',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
# First, create new columns for percentages
tracts_age['PCT_18_to_24'] = tracts_age['TotalPop 18 to 24']/tracts_age['Total Population']*100
tracts_age['PCT_25_to_34'] = tracts_age['TotalPop 25 to 34']/tracts_age['Total Population']*100
tracts_age['PCT_35_to_44'] = tracts_age['TotalPop 35 to 44']/tracts_age['Total Population']*100
```

```python
tracts_age.plot(figsize=(12,10),
                 column='PCT_18_to_24',
                 legend=True, 
                 scheme='equal_interval')
```

```python
tracts_age.plot(figsize=(12,10),
                 column='PCT_25_to_34',
                 legend=True, 
                 scheme='equal_interval')
```

```python
tracts_age.plot(figsize=(12,10),
                 column='PCT_35_to_44',
                 legend=True, 
                 scheme='equal_interval')
```

```python
tracts_age[tracts_age.PCT_25_to_34 > 50].plot(figsize=(12,10))
```
